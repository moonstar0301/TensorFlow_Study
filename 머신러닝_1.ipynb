텐서플로2(TensorFlow 2)는 구글이 개발한 딥 러닝 프레임워크 중 하나입니다. 딥 러닝 모델의 구축, 학습 및 배포를 위한 도구들을 제공하며, 엔드 투 엔드(end-to-end) 딥 러닝 솔루션을 구축할 수 있도록 설계되었습니다.

텐서플로2는 높은 수준의 추상화(abstraction)를 제공하여 개발자가 더욱 쉽게 딥 러닝 모델을 구현할 수 있도록 돕습니다. 즉, 다양한 수준의 API를 제공하여 개발자가 최소한의 코드로 딥 러닝 모델을 구현할 수 있도록 지원합니다.

또한, 텐서플로2는 즉시 실행(eager execution) 모드를 도입하여 딥 러닝 모델 개발 및 디버깅을 보다 쉽게 할 수 있도록 했습니다. 이전 버전의 텐서플로에서는 그래프 실행 모드(graph execution mode)를 사용하여 딥 러닝 모델을 실행해야 했으나, 텐서플로2에서는 즉시 실행 모드를 사용하여 디버깅 과정을 보다 쉽게 할 수 있습니다.

또한, 텐서플로2는 케라스(Keras) API를 기본으로 사용하여 딥 러닝 모델을 구현할 수 있도록 지원합니다. 케라스는 사용하기 쉽고 직관적인 API를 제공하여 딥 러닝 모델을 구현하는 데 도움이 되며, 텐서플로2에서는 케라스 API를 강화하여 보다 효율적으로 딥 러닝 모델을 구현할 수 있도록 했습니다.

마지막으로, 텐서플로2는 분산 학습(distributed training)을 지원하여 대용량 데이터셋에서 빠르게 딥 러닝 모델을 학습할 수 있도록 합니다. 이를 위해 다중 GPU 및 TPU(Tensor Processing Unit)를 지원하며, 분산 학습 API를 제공하여 개발자가 더욱 쉽게 분산 학습을 구현할 수 있도록 합니다.

"텐서를 만든다"는 의미는 데이터를 다차원 배열 형태로 변환하는 것을 의미합니다. 텐서(Tensor)는 수학적인 개념으로, 스칼라(scalar), 벡터(vector), 행렬(matrix)과 같은 다양한 차원의 배열을 표현하는데 사용됩니다. 딥 러닝에서는 데이터를 텐서 형태로 변환하여 처리하고, 딥 러닝 모델의 입력 및 출력으로 텐서를 사용합니다.

따라서, "텐서를 만든다"는 의미는 데이터를 텐서 형태로 변환하는 과정을 의미합니다. 예를 들어, 이미지 데이터를 딥 러닝 모델에 입력하기 위해서는 이미지를 수치 데이터로 변환하고, 이를 다차원 배열 형태인 텐서로 변환해야 합니다. 이 과정에서, 이미지 데이터를 수치 데이터로 변환하고, 이를 텐서로 변환하는 과정이 포함됩니다.

```python
# Basic Tensor Operations
from __future__ import print_function
# __future__를 통해 상위 버전의 기능을 쓸 수 있도록 하였습니다
import tensorflow as tf
```

```python
#define tensor constants. 
a = tf.constant(2)
b = tf.constant(3)
c = tf.constant(5)
```

```python
#Various tensor operations
#Note : Tensors also support python operations

#텐서의 인스턴스를 사용해서 temsor operatiions 사용
add = tf.add(a, b)
sub = tf.subtract(a, b)
mul = tf.multiply(a, b)
div = tf.divide(a, b)

#Access tensors value
print("add = ", add.numpy())
print("sub = ", sub.numpy())
print("mul = ", mul.numpy())
print("div = ", div.numpy())
```

```python
#some more operations
mean = tf.reduce_mean([a, b, c])
sum = tf.reduce_sum([a, b, c])
#차원을 줄이면서 연산해서 reduce_mean임
#텐서플로우에서 2.와 같은 부동소수점 표현 방식을 사용하여,
#딥 러닝 모델에서 필요한 수치 연산을 정확하게 처리할 수 있습니다.

#Access tensors value.
print("mean = ", mean.numpy())
print("sum = ", sum.numpy())
```

```python
# Linear Regression
# Linear Regression implementation with TensorFlow v2 library

from __future__ import absolute_import, division, print_function
```

```python
import tensorflow as tf
import numpy as np
rng = np.random
```

```python
#parameters.
learning_rate = 0.01
training_steps = 1000
display_step = 50
```

```python
# Training Data.
X = np.array([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
              7.042,10.791,5.313,7.997,5.654,9.27,3.1])
Y = np.array([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
              2.827,3.465,1.65,2.904,2.42,2.94,1.3])
```

```python
#Weight and Bias, initialized randomly.
W = tf.Variable(rng.randn(), name = "weight")
b = tf.Variable(rng.randn(), name = "bias")

#Linear regression(Wx + b)
def linear_regression(x):
  return W*x + b

#Mean square error
def mean_square(y_pred, y_true):
  return tf.reduce_mean(tf.square(y_pred - y_true))

#Stochastic Gradient Descent Optimizer(확률적 경사하강법)
optimizer = tf.optimizers.SGD(learning_rate)
```

```python
#Optimization process
def run_optimization():
  #Wrap computation inside a GradientTape for automatic differentiation.
  with tf.GradientTape() as g:
    pred = linear_regression(X)
    loss = mean_square(pred, Y)

    #Compute gradients.
    gradients = g.gradient(loss, [W, b])

    #Update W and b following gradients
    optimizer.apply_gradients(zip(gradients, [W, b]))
```

```python
#Run training for the given number of steps
for step in range(1, training_steps + 1):
  #Run the optimization to update W and b values
  run_optimization()

  if step%display_step == 0:
    pred = linear_regression(X)
    loss = mean_square(pred, Y)
    print("step: %i, loss: %f, W: %f, b: %f" % (step, loss, W.numpy(), b.numpy()))
```

```python
import matplotlib.pyplot as plt
```

```python
# Graphic display
plt.plot(X, Y, 'ro', label='Original data')
plt.plot(X, np.array(W * X + b), label='Fitted line')
plt.legend()
plt.show()
```

다중 입력에 대한 실습

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
# 중간 고사, 기말 고사, 가산점 점수
X = np.array([[70,85,11], [71,89,18], [50,80,20], [99,20,10], [50,10,10]]) 
y = np.array([73, 82 ,72, 57, 34]) # 최종 성적

#Sequential()은 Keras 모델을 생성하는 함수입니다. 이 함수를 사용해서
#레이어를 차례로 추가하여 딥 러닝 모델을 만들수 있습니다.

#입력 뉴런이 3개, 출력 뉴런이 1개인 fully connected layer를 생성하고
#이를 선형 활성화 함수로 연결하여 딥 러닝 모델을 생성하는 코드입니다.

model = Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))

#model.add()는 새로운 레이어를 모델에 추가
```

```python
#model.compile() 메서드는 Keras 모델의 컴파일 단계로,
#모델을 학습하기 전에 최적화 알고리즘, 손실 함수, 평가 지표 등을 설정하는 과정입니다.

sgd = optimizers.SGD(lr=0.0001)
model.compile(optimizer=sgd, loss='mse', metrics=['mse'])
model.fit(X, y, epochs=100)
```

```python
print(model.predict(X))
```

```python
X_test = np.array([[20,99,10], [40,50,20]])
print(model.predict(X_test))
```
